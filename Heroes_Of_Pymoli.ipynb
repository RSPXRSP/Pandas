
## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
pd.options.display.float_format = '{:.2f}'.format

data_gender = pd.DataFrame(df.loc[:,'Gender'])
data_price = pd.DataFrame(df.loc[:,'Price'])

Gender_Analysis = pd.concat([data_gender, data_price], axis=1)

Gen_Pur_Count = Gender_Analysis['Gender'].value_counts()
Gen_Pur_Count = pd.DataFrame(Gen_Pur_Count)
Gen_Pur_Count.index = ('Male','Female','Other / Non-Disclosed')

Gen_Pur_Ave = Gender_Analysis.groupby(['Gender']).mean()
Men_Average = round(Gen_Pur_Ave.loc['Male','Price'],2)
Female_Average = round(Gen_Pur_Ave.loc['Female','Price'],2)
Other_Average = round(Gen_Pur_Ave.loc['Other / Non-Disclosed','Price'],2)
Gen_Pur_Ave = np.concatenate([Men_Average, Female_Average, Other_Average], axis=None)
Gen_Pur_Ave = np.array(Gen_Pur_Ave, dtype=float).tolist()

Gen_Pur_Ave = pd.DataFrame(Gen_Pur_Ave)
Gen_Pur_Ave = pd.concat([Gen_Pur_Ave], axis=1)
Gen_Pur_Ave.columns = ["Average Purchase Price"]
Gen_Pur_Ave.index = ('Male','Female','Other / Non-Disclosed')

Gen_Pur_Total = Gender_Analysis.groupby(['Gender']).sum()
Men_Total = round(Gen_Pur_Total.loc['Male','Price'],2)
Female_Total = round(Gen_Pur_Total.loc['Female','Price'],2)
Other_Total = round(Gen_Pur_Total.loc['Other / Non-Disclosed','Price'],2)

Gen_Pur_Total = np.concatenate([Men_Total, Female_Total, Other_Total], axis=None)
Gen_Pur_Total = np.array(Gen_Pur_Total).tolist()

Gen_Pur_Total = pd.DataFrame(Gen_Pur_Total)
Gen_Pur_Total = pd.concat([Gen_Pur_Total], axis=1)
Gen_Pur_Total.columns = ["Total Purchase Value"]
Gen_Pur_Total.index = ('Male','Female','Other / Non-Disclosed')

Gender_Person_Count = df.groupby('SN').filter(lambda g: g.Gender.nunique() == 1)
Gender_Person_Count = Gender_Person_Count.groupby('Gender').SN.nunique()

Male_SN_Count = Gender_Person_Count['Male'].sum()
Female_SN_Count = Gender_Person_Count['Female'].sum()
Other_SN_Count = Gender_Person_Count['Other / Non-Disclosed'].sum()

Ave_Pur_SN_Male = round(Men_Total / Male_SN_Count,2)
Ave_Pur_SN_Female = round(Female_Total / Female_SN_Count,2)
Ave_Pur_SN_Other = round(Other_Total / Other_SN_Count,2)

Ave_Pur_Person = np.concatenate([Ave_Pur_SN_Male, Ave_Pur_SN_Female, Ave_Pur_SN_Other], axis=None)
Ave_Pur_Person = np.array(Ave_Pur_Person).tolist()

Ave_Pur_Person = pd.DataFrame(Ave_Pur_Person)
Ave_Pur_Person = pd.concat([Ave_Pur_Person], axis=1)
Ave_Pur_Person.columns = ["Avg Total Purchase per Person"]
Ave_Pur_Person.index = ('Male','Female','Other / Non-Disclosed')

Gender_Summary = pd.concat([Gen_Pur_Count, Gen_Pur_Ave, Gen_Pur_Total, Ave_Pur_Person], axis=1)
Gender_Summary = pd.DataFrame(Gender_Summary)
Gender_Summary
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
data_Age = pd.DataFrame(df.loc[:,'Age'])
data_SN = pd.DataFrame(df.loc[:,'SN'])

Age_Dem = pd.concat([data_Age, data_SN], axis=1)
Age_Dem = pd.DataFrame.drop_duplicates(Age_Dem)
Age_Total = Age_Dem.loc[:,"SN"].count()

labels = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '+40']
bins = [0,9,14,19,24,29,34,39,100]
bins = pd.cut(Age_Dem['Age'], bins=bins)
Age_Count = Age_Dem.groupby(bins)['Age'].count()
Age_Mean = Age_Count / Age_Total * 100
Age_Mean = Age_Mean.map('{:,.2f}%'.format)
Age_Summary = pd.concat([Age_Count, Age_Mean], axis=1)
Age_Summary.columns = ['Total Count','Percentage of Players']
Age_Summary.index = labels
Age_Summary
```
